```
import requests
import json
import csv

from google.colab import drive
drive.mount('/gdrive')

id_firiem = set() 
firmy = set() 

def sprocesuj (id_firmy):
  idecka_zavierok,ico_firmy = najdi_zavierky(id_firmy)
  try:
    druha = najdi_par(idecka_zavierok)
  except:
    druha = 0
  print (ico_firmy, druha)

  if druha > 0:
    vynos2018 = vrat_vynosy(druha)
  else:
    vynos2018 = -10000000
  return vynos2018

def najdi_zavierky(id_firmy):
  jednotka_url = 'http://www.registeruz.sk/cruz-public/api/uctovna-jednotka?id='+str(id_firmy)
  y = requests.get(jednotka_url)
  odpovedy = json.loads(y.text)
  
  try:
    if odpovedy['kraj'] == 'SK032' or odpovedy['kraj'] == 'SK041' or odpovedy['kraj'] == 'SK042':
        return odpovedy["idUctovnychZavierok"],odpovedy["ico"]
    else:
        return 0,0
  except:
    return 0,0

def najdi_par(idzavierky):
  posledna = max(idzavierky)
  zavierka_url = 'http://www.registeruz.sk/cruz-public/api/uctovna-zavierka?id='+str(posledna)
  z = requests.get(zavierka_url)
  odpovedz = json.loads(z.text)
  if odpovedz['obdobieOd'] == '2019-01':
    posledna = min(odpovedz['idUctovnychVykazov'])
  else:
    posledna = 0
  return posledna

def vrat_vynosy(idvykazu):
  try:
    vykaz_url = 'http://www.registeruz.sk/cruz-public/api/uctovny-vykaz?id='+str(idvykazu)
    v = requests.get(vykaz_url)
    odpovedv = json.loads(v.text)
    level1 = odpovedv['obsah']['tabulky']
    level2 = level1[2]
    level3 = level2['data']
    level4 = level3[0]
  except:
    level4 = 0
  return level4



zostava = 351793
id_od = 1

icos = []
vynosy = {}

while zostava == 351793:
  id_url = 'http://www.registeruz.sk/cruz-public/api/uctovne-jednotky?zmenene-od=1990-01-01&max-zaznamov=1000&pravna-forma=112&pokracovat-za-id='+str(id_od)
  x = requests.get(id_url)
  odpovedx = json.loads(x.text)  
  zostava = odpovedx["pocetZostavajucichId"]
  idecka_firiem = odpovedx["id"]
  for id_firmy in idecka_firiem:
    firmy.add(id_firmy)

  id_od = max(idecka_firiem)




#print(icos)

#icos = [46270469]

output_file = "/gdrive/My Drive/vystup20200527_1.csv" 
with open(output_file, 'w', encoding='iso8859_2', newline='') as scat:
    csv.excel.delimiter=';'
    wr = csv.writer(scat, quoting=csv.QUOTE_NONE ,escapechar='\\', dialect=csv.excel)
    wr.writerow(["ICO","2019"])

    for firma in firmy:

      try:
          vynos2018 = sprocesuj (firma)
          if vynos2018 != -10000000:
            vynosy[ico] = vynos2018
            wr.writerow([ico,vynos2018])
    
          else:
            print (ico,"nefunguje")
      except:
        continue










```
